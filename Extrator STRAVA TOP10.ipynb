```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from webdriver_manager.firefox import GeckoDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import TimeoutException
import pandas as pd
import time
import pandas as pd

custom_path = r'C:\Users\Matheus\.wdm\drivers\chromedriver\win64\115.0.5790.171\chromedriver.exe'
driver = webdriver.Chrome(executable_path=custom_path)

driver.get("https://www.strava.com")

# Criar um DataFrame para armazenar os dados
colunas = ["Segmento", "UF", "Data", "Distancia", "Ganho de Elevacao", "Inclinação média",
           "Elev. mais baixa", "Elev. mais alta", "Diferença de elev", "Nº de Tentativas",
           "Tempo", "Posicao", "Tempo 1º", "Tempo 10", "Veloc. Média"]

strava_top10 = pd.DataFrame(columns=colunas)
```

```python
# Esperar até aparecer o Xpath
wait = WebDriverWait(driver, 10)
wait1 = WebDriverWait(driver, 20)

# Definir a variável atividades_processadas no início do código
atividades_processadas = 0

# Número total de atividades e número de atividades por página
NUMERO_TOTAL_DE_ATIVIDADES = 180
ATIVIDADES_POR_PAGINA = 20
pagina_atual = 0
atividade_index = 0

# Loop para iterar por diferentes índices de atividades
for atividade_index in range(atividade_index, NUMERO_TOTAL_DE_ATIVIDADES + 1):
    if atividade_index > 0 and atividade_index % ATIVIDADES_POR_PAGINA == 0:
        # Calcular quantas vezes clicar no botão "Próximo" para voltar à próxima página
        vezes_clicar = atividade_index // ATIVIDADES_POR_PAGINA

        try:
            if vezes_clicar > 0:
                # Clicar no botão "Próximo" uma vez para ir para a próxima página
                click_next = wait.until(EC.element_to_be_clickable((By.XPATH, f'/html/body/div[1]/div[3]/div[3]/div[1]/div/nav/ul/li[last()]/a')))
                click_next.click()
                wait.until(EC.staleness_of(click_next))

                # Atualizar o número da página atual
                pagina_atual += 1

        except TimeoutException:
            print(f"Tempo limite atingido ao tentar clicar em 'Próximo' na página {pagina_atual}.")

    # Clique para abrir a página da atividade
    click_atividade = wait1.until(EC.visibility_of_element_located((By.XPATH, f'/html/body/div[1]/div[3]/div[3]/div[1]/div/table/tbody/tr[{(atividade_index % ATIVIDADES_POR_PAGINA) + 1}]/td[2]/a')))
    click_atividade.click()
    time.sleep(2)

    # Nome do segmento
    atividade_element = driver.find_element(By.XPATH, "//*[@id='js-full-name']")
    nome_segmento = atividade_element.text if atividade_element else "NaN"
    # Adicionar a informação à coluna "Segmento" no DataFrame existente
    strava_top10.loc[atividades_processadas + 1, "Segmento"] = nome_segmento

    # UF
    uf_element = driver.find_element(By.XPATH, "/html/body/div[2]/div[1]/div/div[3]/div")
    uf_segmento = uf_element.text if uf_element else "NaN"
    # Adicionar a informação à coluna "UF" no DataFrame existente
    strava_top10.loc[atividades_processadas + 1, "UF"] = uf_segmento

    # Data
    data_element = driver.find_element(By.XPATH, "/html/body/div[2]/div[2]/div[2]/div[1]/ul/li[1]/div[1]/div/div[2]/div[2]/div/a")
    data_segmento = data_element.text if data_element else "NaN"
    # Adicionar a informação à coluna "Data" no DataFrame existente
    strava_top10.loc[atividades_processadas + 1, "Data"] = data_segmento

    # Distância
    distancia_element = driver.find_element(By.XPATH, '//*[@id="segment"]/div/div[3]/ul/li[1]/div/b')
    distancia_segmento = distancia_element.text if distancia_element else "NaN"
    # Adicionar a informação à coluna "Distancia" no DataFrame existente
    strava_top10.loc[atividades_processadas + 1, "Distancia"] = distancia_segmento

    # Ganho elevação
    elevacao_element = driver.find_element(By.XPATH, '//*[@id="segment"]/div/div[3]/ul/li[2]/div/b')
    elevacao_segmento = elevacao_element.text if elevacao_element else "NaN"
    # Adicionar a informação à coluna "Ganho de Elevacao" no DataFrame existente
    strava_top10.loc[atividades_processadas + 1, "Ganho de Elevacao"] = elevacao_segmento

    # Inclinação média
    inclinacao_element = driver.find_element(By.XPATH, '//*[@id="segment"]/div/div[3]/ul/li[3]/div/b')
    inclinacao_segmento = inclinacao_element.text if inclinacao_element else "NaN"
    # Adicionar a informação à coluna "Inclinação média" no DataFrame existente
    strava_top10.loc[atividades_processadas + 1, "Inclinação média"] = inclinacao_segmento

    # Elev. mais baixa
    elevacao1_element = driver.find_element(By.XPATH, '//*[@id="segment"]/div/div[3]/ul/li[4]/div/b')
    elevacao1_segmento = elevacao1_element.text if elevacao1_element else "NaN"
    # Adicionar a informação à coluna "Elev. mais baixa" no DataFrame existente
    strava_top10.loc[atividades_processadas + 1, "Elev. mais baixa"] = elevacao1_segmento

    # Elev. mais alta
    elevacao1_element = driver.find_element(By.XPATH, '//*[@id="segment"]/div/div[3]/ul/li[5]/div/b')
    elevacao1_segmento = elevacao1_element.text if elevacao1_element else "NaN"
    # Adicionar a informação à coluna "Elev. mais alta" no DataFrame existente
    strava_top10.loc[atividades_processadas + 1, "Elev. mais alta"] = elevacao1_segmento

    # Diferença de elev.
    elevacao2_element = driver.find_element(By.XPATH, '//*[@id="segment"]/div/div[3]/ul/li[6]/div/b')
    elevacao2_segmento = elevacao2_element.text if elevacao2_element else "NaN"
    # Adicionar a informação à coluna "Diferença de elev" no DataFrame existente
    strava_top10.loc[atividades_processadas + 1, "Diferença de elev"] = elevacao2_segmento

    # Nº de Tentativas
    tentativas_element = driver.find_element(By.XPATH, '/html/body/div[2]/div[2]/div[2]/div[1]/ul/li[1]/div[1]/div/div[2]/div[2]/ul/li[2]/div/div[2]')
    tentativas_segmento = tentativas_element.text if tentativas_element else "NaN"
    # Adicionar a informação à coluna "Nº de Tentativas" no DataFrame existente
    strava_top10.loc[atividades_processadas + 1, "Nº de Tentativas"] = tentativas_segmento

    # Tempo
    tempo_element = driver.find_element(By.XPATH, '/html/body/div[2]/div[2]/div[2]/div[1]/ul/li[1]/div[1]/div/div[2]/div[2]/ul/li[1]/div/div[2]')
    tempo_segmento = tempo_element.text if tempo_element else "NaN"
    # Adicionar a informação à coluna "Tempo" no DataFrame existente
    strava_top10.loc[atividades_processadas + 1, "Tempo"] = tempo_segmento

    # Posicao
    posicao_element = driver.find_element(By.XPATH, '//*[@id="results"]/table/tbody/tr[2]')
    posicao_segmento = posicao_element.text if posicao_element else "NaN"
    # Adicionar a informação à coluna "Posicao" no DataFrame existente
    strava_top10.loc[atividades_processadas + 1, "Posicao"] = posicao_segmento

    # Tempo 1º colocado
    tempo1_element = driver.find_element(By.XPATH, '//*[@id="results"]/table/tbody/tr[1]/td[7]')
    tempo1_segmento = tempo1_element.text if tempo1_element else "NaN"
    # Adicionar a informação à coluna "Tempo 1º" no DataFrame existente
    strava_top10.loc[atividades_processadas + 1, "Tempo 1º"] = tempo1_segmento
    
    # Tempo 2 colocado
    try:
        tempo2_element = driver.find_element(By.XPATH, '//*[@id="results"]/table/tbody/tr[2]')
        tempo2_segmento = tempo2_element.text
    except NoSuchElementException:
        tempo2_segmento = "NaN"
    # Adicionar a informação à coluna "Tempo 10" no DataFrame existente
    strava_top10.loc[atividades_processadas + 1, "Tempo 2"] = tempo2_segmento
    
# Tempo 3 colocado
    try:
        tempo3_element = driver.find_element(By.XPATH, '//*[@id="results"]/table/tbody/tr[3]')
        tempo3_segmento = tempo3_element.text
    except NoSuchElementException:
        tempo3_segmento = "NaN"
    # Adicionar a informação à coluna "Tempo 10" no DataFrame existente
    strava_top10.loc[atividades_processadas + 1, "Tempo 3"] = tempo3_segmento
    
    
# Tempo 4 colocado
    try:
        tempo4_element = driver.find_element(By.XPATH, '//*[@id="results"]/table/tbody/tr[4]')
        tempo4_segmento = tempo4_element.text
    except NoSuchElementException:
        tempo4_segmento = "NaN"
    # Adicionar a informação à coluna "Tempo 10" no DataFrame existente
    strava_top10.loc[atividades_processadas + 1, "Tempo 4"] = tempo4_segmento
    
# Tempo 5 colocado
    try:
        tempo5_element = driver.find_element(By.XPATH, '//*[@id="results"]/table/tbody/tr[5]')
        tempo5_segmento = tempo5_element.text
    except NoSuchElementException:
        tempo5_segmento = "NaN"
    # Adicionar a informação à coluna "Tempo 10" no DataFrame existente
    strava_top10.loc[atividades_processadas + 1, "Tempo 5"] = tempo5_segmento
    
 # Tempo 6 colocado
    try:
        tempo6_element = driver.find_element(By.XPATH, '//*[@id="results"]/table/tbody/tr[6]')
        tempo6_segmento = tempo6_element.text
    except NoSuchElementException:
        tempo6_segmento = "NaN"
    # Adicionar a informação à coluna "Tempo 10" no DataFrame existente
    strava_top10.loc[atividades_processadas + 1, "Tempo 6"] = tempo6_segmento
    
# Tempo 7 colocado
    try:
        tempo7_element = driver.find_element(By.XPATH, '//*[@id="results"]/table/tbody/tr[7]')
        tempo7_segmento = tempo7_element.text
    except NoSuchElementException:
        tempo7_segmento = "NaN"
    # Adicionar a informação à coluna "Tempo 10" no DataFrame existente
    strava_top10.loc[atividades_processadas + 1, "Tempo 7"] = tempo7_segmento
    
# Tempo 8 colocado
    try:
        tempo8_element = driver.find_element(By.XPATH, '//*[@id="results"]/table/tbody/tr[8]')
        tempo8_segmento = tempo8_element.text
    except NoSuchElementException:
        tempo8_segmento = "NaN"
    # Adicionar a informação à coluna "Tempo 10" no DataFrame existente
    strava_top10.loc[atividades_processadas + 1, "Tempo 8"] = tempo8_segmento
    
# Tempo 9 colocado
    try:
        tempo9_element = driver.find_element(By.XPATH, '//*[@id="results"]/table/tbody/tr[9]')
        tempo9_segmento = tempo9_element.text
    except NoSuchElementException:
        tempo9_segmento = "NaN"
    # Adicionar a informação à coluna "Tempo 10" no DataFrame existente
    strava_top10.loc[atividades_processadas + 1, "Tempo 9"] = tempo9_segmento
    

    # Tempo 10 colocado
    try:
        tempo10_element = driver.find_element(By.XPATH, '//*[@id="results"]/table/tbody/tr[10]/td[7]')
        tempo10_segmento = tempo10_element.text
    except NoSuchElementException:
        tempo10_segmento = "NaN"
    # Adicionar a informação à coluna "Tempo 10" no DataFrame existente
    strava_top10.loc[atividades_processadas + 1, "Tempo 10"] = tempo10_segmento

    # Veloc. Média
    vmedia_element = driver.find_element(By.XPATH, '//*[@id="results"]/table/tbody/tr[2]/td[4]')
    vmedia_segmento = vmedia_element.text if vmedia_element else "NaN"
    # Adicionar a informação à coluna "Veloc. Média" no DataFrame existente
    strava_top10.loc[atividades_processadas + 1, "Veloc. Média"] = vmedia_segmento

    # Atualiza o contador de atividades processadas
    atividades_processadas += 1

    # Voltar à página anterior
    driver.back()
    time.sleep(3)

# Fechar o navegador
driver.quit()

# Salvar o DataFrame em um arquivo CSV
strava_top10.to_csv('strava_top10.csv', index=False)
```
